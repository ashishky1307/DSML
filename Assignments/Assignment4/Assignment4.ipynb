```python
from sklearn.cluster import KMeans
import warnings
warnings.filterwarnings("ignore")
```

```python
cluster=[]
for k in range(1,11):
  kmean=KMeans(n_clusters=k).fit(df)
  cluster.append(kmean.inertia_)
```

```python
plt.plot(range(1,11),cluster,marker="o")
plt.xlabel('Cluster')
plt.ylabel('Inertia')
plt.show()
```

```python
ka = KMeans(n_clusters=5).fit(df)
df["Labels"]=ka.labels_
df.head()
```

```python
centers = np.array(ka.cluster_centers_)
```

```python
colors = {0:'red',1:'green',2:'orange',3:'yellow',4:'pink'}
plt.figure(figsize=(12,8))
plt.scatter(df['Annual Income (k$)'],df['Spending Score (1-100)'],linewidths=0.5,color=df['Labels'].map(colors))
plt.title('kmeans clustering')
plt.show()
```

```python
colors = {0:'red',1:'green',2:'orange',3:'yellow',4:'pink'}
plt.figure(figsize=(12,8))
plt.scatter(df['Annual Income (k$)'],df['Spending Score (1-100)'],linewidths=0.5,color=df['Labels'].map(colors))
plt.scatter(ka.cluster_centers_[:,2],ka.cluster_centers_[:,3],s=100,c='black',label='Centroids',marker='x')
plt.title('kmeans clustering')
plt.show()
```
