```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
sample=pd.read_csv("/content/temperatures.csv")
```

```python
sample.head()
```

```python
sample.describe()
```

```python
sample.plot(x="YEAR",y="JAN", kind="scatter")
```

```python
Array=np.array("JAN,FEB,MAR,APR,,MAY,JUN,JUL,AUG,SEP,OCT,NOV,DEC")
valid=True
while valid:
  month = input("Enter a month ").upper()
  if month in Array:
    valid = True
  else:
     valid=False

```

```python
x=sample["YEAR"].values.reshape(-1,1)
y=sample[month].values.reshape(-1,1)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=40)
```

```python
data=LinearRegression()
data.fit(x_train,y_train)
```

```python
y_pred=data.predict(x_test)
```

```python
print(data.intercept_)
print(data.coef_)
```

```python
plt.scatter(x_test,y_test,color="blue")
plt.plot(x_test,y_pred, color="green")
plt.show()
```

```python
print("Mean Absolute Error: ", metrics.mean_absolute_error(y_test, y_pred))
print("Mean Squared Error: ", metrics.mean_squared_error(y_test, y_pred))
print("R2 Score: ", metrics.r2_score(y_test, y_pred))
```
